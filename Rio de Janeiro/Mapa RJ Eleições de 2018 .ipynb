```python
# vamos carregar o arquivo com resultado da eleiçao no Rio de Janeiro dividido por municipio/zona

df_votacao = pd.read_csv('dados_eleicao/votacao_candidato_munzona_2018_RJ.csv', sep=';', encoding='latin1')

df_votacao.head()
```

```python
# Visualizando o df transposto
df_votacao.head().T
```

```python
df_votacao['NR_TURNO'].value_counts()
```

```python
df_votacao['NM_CANDIDATO'].value_counts()
```

```python
# Separa os dados pois precisamos apenas dos dados de 2 turno
df_votacao = df_votacao[df_votacao['NR_TURNO'] == 2]
```

```python
df_votacao.sample(4).T
```

```python
df_votacao.info()
```

```python
# VAmos verificar quantos municipios unicos existem no conjunto de dados
# vamos pegar  por codigo do municipio

df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# os dados estao separados por municipio/zona
#Municipios peq normalmente tem apenas 1 zona eleitoral
# mas municipios maiores podem ter varias zonas eleitorais

# vms verificar DUQUE DE CAXIAS
df_votacao[df_votacao['NM_MUNICIPIO'] == 'DUQUE DE CAXIAS']
```

```python
# Eliminar algumas colunas no df  q n sera utilizada
df_votacao = df_votacao[['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS']]

df_votacao.head()
```

```python
# Precisamos agregar os dados por municipio e candidato
df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum()
```

```python
# Fazer o groupby, reindexar o indice vai salvar em um df
df_votacao = df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum().reset_index()
df_votacao.sample(4)
```

```python
df_votacao.info()
```

```python
# vms verificar a cidade 'DUQUE DE CAXIAS'
df_votacao[df_votacao['NM_MUNICIPIO'] == 'DUQUE DE CAXIAS']
```

```python
# ordena o df pela qtd de vts e dps dx apenas o primeiro
# Resgistro de cada municipio

df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False).head(10)
```

```python
# vamos ordenar 
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False, inplace=True)

#dropa as linhas duplicadas, mantendo a primeira
df_votacao.drop_duplicates(subset='CD_MUNICIPIO', keep='first', inplace=True)

df_votacao.head(10)
```

```python
# vms verificar novamente a cidade 'DUQUE DE CAXIAS'
df_votacao[df_votacao['NM_MUNICIPIO'] == 'DUQUE DE CAXIAS']
```

```python
# vamos criar uma coluna representando o resultado da eleiçao naquele municipio
# sendo que o resultado e True se o candidato WILSON WITZEL venceu naquele municipio
# e False em caso contrario
df_votacao['RESULTADO'] = df_votacao['NR_CANDIDATO'].apply(lambda x: x == 20)
```

```python
# Verificar o resultado
df_votacao.head(10)
```

```python
# Mostrando alguns valores do df do geopandas
df_mapa.sample(3)
```

```python
# recuperar a tabela de correspondencia entre ibge e tse
#Direto do github do Estadão
link = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'
df_equivalencia = pd.read_csv(link)
df_equivalencia.head()
```

```python
df_equivalencia.info()
```

```python
df_equivalencia = df_equivalencia[df_equivalencia['UF'] == 'RJ']
```

```python
df_equivalencia.head()
```

```python
df_equivalencia.info()
```

```python
#Vamos preparar os dataframes
# vms reindexar o df de votaçao pelo codigo do TSE
df_vot_novo = df_votacao.copy()

df_vot_novo.set_index('CD_MUNICIPIO', drop=False, inplace=True)

df_vot_novo.head()
```

```python
# vms reindexar o df de Equivalencia pelo o codigo do TSE

df_equi_novo = df_equivalencia.copy()

df_equi_novo.set_index('COD_TSE', drop=False, inplace=True)

df_equi_novo.head()
```

```python
# Criando um novo dataf com base no merge de votaçao e Equivalencia

df_vot_equi = df_vot_novo.join(df_equi_novo)

df_vot_equi.head()
```

```python
# vms junta o data frame do Geopandas com o df votaçao/equivalencia
```

```python
df_vot_equi.info()
```

```python
df_mapa.info()
```

```python
# Transforma a coluna de codigo de IBGE do df votaçao/equivalencia
df_vot_equi['GEOCOD_IBGE'] = df_vot_equi['GEOCOD_IBGE'].astype(str)

df_vot_equi.info()
```

```python
# vms unificar os indices

# Mudando indice do df mapa
df_mapa.set_index('CD_GEOCMU', drop=False, inplace=True)

df_mapa.head(2)
```

```python
# muda o indice do df votaçao/equivalencia
df_vot_equi.set_index('GEOCOD_IBGE', drop=False, inplace=True)
df_vot_equi.head(2)
```

```python
# vms unificar o dataframe do mapa com dataframe de votaçao/equivalencia
df_mapa_novo = df_mapa.join(df_vot_equi)

df_mapa_novo.head().T
```

```python
# verificando o tam do mapa novo
df_mapa_novo.info()
```

```python
#Matar a coluna ajuste
df_mapa_novo.drop(columns=['AJUSTE'], inplace=True)
```

```python
df_mapa_novo.info()
```

```python
# vamos desenhar o mapa

#Criar a referencia para o codigo de cor(False/0 = yellow True/1 = green)
from matplotlib.colors import ListedColormap

cmap = ListedColormap(['yellow', 'green'])

#Desenhar o mapa
fig, ax = plt.subplots(1, figsize=(12, 12))

ax = df_mapa_novo.plot(column='NM_URNA_CANDIDATO', cmap=cmap, legend=True, linewidth=0.1, ax=ax, edgecolor='grey')

# Esconder os eixos
ax.set_axis_off()

#Colocando titulo
ax.set_title('Resultado Eleiçao 2018 por Municipio / 2 Turno RJ', fontdict={'fontsize': '25', 'fontweight': '3'})

#Criando uma nota rodape
ax.annotate('Fonte: IBGE, TSE, 2019; Jornal o Estado de São Paulo, 2019',
    xy=(0.1, 0.08), xycoords='figure fraction', horizontalalignment='left',
    verticalalignment='top', fontsize=10, color='#555555')

#Exibir

plt.show()
```
